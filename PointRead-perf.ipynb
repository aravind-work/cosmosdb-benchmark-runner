```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

# This notebook visualizes the perf of sync, v2, v3, and v4 SDKs on f4 Ubuntu Linux Azure VM
Adobe has two workflows point-read and read-many query

## Point Read Throughput on 4 CPU cores VM (F4s) 
(v4.0.1-beta.4, v4.0.1-beta.3, v4.0.1-beta.2, v4.0.1-beta.1)

```python
point_read_f4s = pd.read_csv('notebooks/20200605/f4/benchmark-results.csv', skipinitialspace=True)
sns.lineplot(x='ThreadCount', y ='Throughput(ops/s)', hue='OpName', data=(point_read_f4s[(point_read_f4s.ThreadCount < 300) & (point_read_f4s.ThreadCount > 0) & (point_read_f4s.OpName > 'PointRead-v4')]))
```
