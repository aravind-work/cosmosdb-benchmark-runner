```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

# This notebook visualizes the throughput perf of sync, v3, and v4 SDKs.
Adobe has two workflows point-read and read-many query

## Point Read Throughput on 2 CPU cores VM (F2s)

```python
point_read_f2s = pd.read_csv('notebooks/20200214-adobe/f2s/point-read.csv', skipinitialspace=True)
sns.lineplot(x='ThreadCount', y ='Throughput(ops/s)', hue='OpName', data=(point_read_f2s[point_read_f2s.ThreadCount < 300]))
```

## Point Read Throughput on 4 CPU cores VM (F4s)

```python
point_read_f4s = pd.read_csv('notebooks/20200214-adobe/f4s/point-read.csv', skipinitialspace=True)
sns.lineplot(x='ThreadCount', y ='Throughput(ops/s)', hue='OpName', data=(point_read_f4s[(point_read_f4s.ThreadCount < 500) & (point_read_f4s.ThreadCount > 0)]))
```
